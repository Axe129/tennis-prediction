```python
#testing - 
player_dict = {}
surfaces = ['Hard', 'Grass', 'Clay']
global player_count
player_list = []
player_count = 0

def fill_dictionary(player1, year, ATP, tournament, player1_sets_won, total_sets_played, surface, winner, date):
    global player_count
    if player1 not in player_dict:
        print("Adding player: ", player1)
        player_list.append(player1)
        player_count = player_count + 1
        player_dict[player1] = {}
    
    if year not in player_dict[player1]:
        player_dict[player1][year] = {'ATPs':[], 'tournaments': [], 'dates':[], 'sets_played':{'Total':[], 'Hard':[], 'Grass':[], 'Clay':[]}, 'sets_won':{'Total':[], 'Hard':[], 'Grass':[], 'Clay':[]}, 'matches_played':{'Total':[], 'Hard':[], 'Grass':[], 'Clay':[]}, 'matches_won':{'Total':[], 'Hard':[], 'Grass':[], 'Clay':[]}}
    
    if tournament not in player_dict[player1][year]['tournaments']:
        
        player_dict[player1][year]['tournaments'].append(tournament)
        player_dict[player1][year]['dates'].append(date)
        player_dict[player1][year]['ATPs'].append(ATP)
        
        if len(player_dict[player1][year]['tournaments']) == 1:
            add_sets_played_total = 0
            add_sets_won_total = 0
            add_matches_played_total = 0
            add_matches_won_total = 0
            
            add_sets_played_surface = 0
            add_sets_won_surface = 0
            add_matches_played_surface = 0
            add_matches_won_surface = 0
        else:
            add_sets_played_total =  player_dict[player1][year]['sets_played']['Total'][-1]
            add_sets_won_total = player_dict[player1][year]['sets_won']['Total'][-1]
            add_matches_played_total = player_dict[player1][year]['matches_played']['Total'][-1]
            add_matches_won_total = player_dict[player1][year]['matches_won']['Total'][-1]
            
            add_sets_played_surface = player_dict[player1][year]['sets_played'][surface][-1]
            add_sets_won_surface = player_dict[player1][year]['sets_won'][surface][-1]
            add_matches_played_surface = player_dict[player1][year]['matches_played'][surface][-1]
            add_matches_won_surface = player_dict[player1][year]['matches_won'][surface][-1]
            
        #player_dict[player1][year]['sets_played'].append(total_sets_played)
        player_dict[player1][year]['sets_played']['Total'].append(total_sets_played + add_sets_played_total)
        player_dict[player1][year]['sets_played'][surface].append(total_sets_played + add_sets_played_surface)

        #player_dict[player1][year]['sets_won'].append(player1_sets_won)
        player_dict[player1][year]['sets_won']['Total'].append(player1_sets_won + add_sets_won_total)
        player_dict[player1][year]['sets_won'][surface].append(player1_sets_won + add_sets_won_surface)


        #player_dict[player1][year]['matches_played'].append(1)
        player_dict[player1][year]['matches_played']['Total'].append(1 + add_matches_played_total)
        player_dict[player1][year]['matches_played'][surface].append(1 + add_matches_played_surface)


        #player_dict[player1][year]['matches_won'].append(1)
        if winner == True:
            player_dict[player1][year]['matches_won']['Total'].append(1 + add_matches_won_total)
            player_dict[player1][year]['matches_won'][surface].append(1 + add_matches_won_surface)
        else:
            player_dict[player1][year]['matches_won']['Total'].append(add_matches_won_total)
            player_dict[player1][year]['matches_won'][surface].append(add_matches_won_surface)
        for surf in surfaces:
            if surf != surface:
                if len(player_dict[player1][year]['tournaments']) == 1:
                    player_dict[player1][year]['sets_played'][surf].append(0)
                    player_dict[player1][year]['sets_won'][surf].append(0)
                    player_dict[player1][year]['matches_played'][surf].append(0)
                    player_dict[player1][year]['matches_won'][surf].append(0)
                else:
                    player_dict[player1][year]['sets_played'][surf].append(player_dict[player1][year]['sets_played'][surf][-1])
                    player_dict[player1][year]['sets_won'][surf].append(player_dict[player1][year]['sets_won'][surf][-1])
                    player_dict[player1][year]['matches_played'][surf].append(player_dict[player1][year]['matches_played'][surf][-1])
                    player_dict[player1][year]['matches_won'][surf].append(player_dict[player1][year]['matches_won'][surf][-1])
    else:
        player_dict[player1][year]['sets_played']['Total'][-1] += total_sets_played
        player_dict[player1][year]['sets_played'][surface][-1] += total_sets_played


        player_dict[player1][year]['sets_won']['Total'][-1] += player1_sets_won
        player_dict[player1][year]['sets_won'][surface][-1] += player1_sets_won


        player_dict[player1][year]['matches_played']['Total'][-1] += 1
        player_dict[player1][year]['matches_played'][surface][-1] += 1

        if winner == True:
            player_dict[player1][year]['matches_won']['Total'][-1] += 1
            player_dict[player1][year]['matches_won'][surface][-1] += 1





for i in range(43141, 0, -1):
    #print('i: ', i)
    if i%5000 == 0:
        print('i: ', i)
    current_match = data.iloc[i]
    year = current_match['Date'].year
    date = current_match['Date']
    player1 = current_match['Winner']
    player1 = player1.strip()
    player2 = current_match['Loser']
    player2 = player2.strip()
    ATP = current_match['ATP']
    tournament = current_match['Tournament']

    player1_sets_won = current_match['Wsets']
    player2_sets_won = current_match['Lsets']
    total_sets_played = player1_sets_won + player2_sets_won

    surface = current_match['Surface']

    fill_dictionary(player1, year, ATP, tournament, player1_sets_won, total_sets_played, surface, True, date)
    fill_dictionary(player2, year, ATP, tournament, player2_sets_won, total_sets_played, surface, False, date)


#print("PLAYER DICT: ")
#player_dict
print("Number of players: ", player_count)
pickle.dump(player_dict, open('player_dict.p', 'wb'))
pickle.dump(player_list, open('player_names.p', 'wb'))
print("Dumped!")
```
